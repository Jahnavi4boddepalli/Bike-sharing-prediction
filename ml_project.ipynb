```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Creating an array to store MSE values
mse_values = []

# Fitting the linear regression model and calculating MSE for different iterations
for i in range(1, len(X)):
    lr = LinearRegression()
    lr.fit(X[:i], y[:i])
    y_pred = lr.predict(X[:i])
    mse = mean_squared_error(y[:i], y_pred)
    mse_values.append(mse)

# Plotting the MSE values
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(X)), mse_values, color='blue')
plt.title('Mean Squared Error (MSE) of Linear Regression')
plt.xlabel('Number of data points')
plt.ylabel('MSE')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel()

# Creating an array to store MSE values
mse_values = []

# Fitting the SVR model and calculating MSE for different dataset sizes
for i in range(1, len(X)):
    svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
    svr.fit(X[:i], y[:i])
    y_pred = svr.predict(X[:i])
    mse = mean_squared_error(y[:i], y_pred)
    mse_values.append(mse)

# Plotting the MSE values
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(X)), mse_values, color='green')
plt.title('Mean Squared Error (MSE) of Support Vector Regression (SVR)')
plt.xlabel('Number of data points')
plt.ylabel('MSE')
plt.grid(True)
plt.show()

# Calculating and printing final MSE and standard mean
final_mse = mse_values[-1]  # Final (last) MSE value
mean_mse = np.mean(mse_values)  # Mean of all MSE values

print(f"Final Mean Squared Error (MSE): {final_mse:.4f}")
print(f"Standard Mean of MSE: {mean_mse:.4f}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
from sklearn.metrics import r2_score, mean_squared_error

# Assuming you have your data loaded in variables X and y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Number of bootstrap iterations
n_bootstrap_iterations = 1000  # You can adjust this number

# Initialize lists to store MSE and R2 scores from each iteration
mse_scores = []
r2_scores = []

for _ in range(n_bootstrap_iterations):
    # Resample the training data with replacement
    X_boot, y_boot = resample(X_train, y_train, replace=True, random_state=42)

    # Create and fit a Linear Regression model
    lr = LinearRegression()
    lr.fit(X_boot, y_boot)

    # Predict on the test set
    y_pred = lr.predict(X_test)

    # Calculate and store the MSE and R2 scores
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    mse_scores.append(mse)
    r2_scores.append(r2)

# Calculate the average MSE and R2 scores over all iterations
average_mse = np.mean(mse_scores)
average_r2 = np.mean(r2_scores)

print("Average Mean Squared Error:", average_mse)
print("Average R2 Score:", average_r2)
```

```python
#applying bootstrap to lasso regression
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
def perform_bootstrap_lasso(X, y, alpha, num_iterations):
    mse_values = []
    coefficients_list = []
    for i in range(num_iterations):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, shuffle=True)
        lasso = Lasso(alpha=alpha)
        lasso.fit(X_train, y_train)
        y_pred = lasso.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mse_values.append(mse)
        coefficients_list.append(lasso.coef_)
    mean_mse = np.mean(mse_values)
    std_mse = np.std(mse_values)
    print(f"Mean of MSE across {num_iterations} iterations for Lasso regression: {mean_mse}")
    print(f"Standard Deviation of MSE for Lasso regression: {std_mse}")
    average_coefficients = np.mean(coefficients_list, axis=0)
    print("Average Lasso Coefficients:", average_coefficients)
# Lasso Regression
perform_bootstrap_lasso(X, y, alpha=0.01, num_iterations=100)
```
