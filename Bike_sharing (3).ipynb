```python
import pandas as pd
a=pd.read_csv('/content/hour.csv')
print(a)
```

```python
a.dtypes
```

```python
x1=a.drop('dteday',axis=1)
#print("features data:")
x2=x1.drop('cnt',axis=1)
print(x2)
```

```python
x=x2.iloc[:,0:14]
y=x2.iloc[:,14:15]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test= train_test_split(x , y, test_size=0.20, random_state=42)
print(y_train,x_train)
lg=LinearRegression()
mm=lg.fit(x_train,y_train)
print(mm.score(x_test,y_test))
print(mm.score(x_train,y_train))
yp=mm.predict(x_test)
#print(accuracy_score(yp,y_test))
```
